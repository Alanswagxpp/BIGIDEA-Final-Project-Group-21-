```python
import pandas as pd
import numpy as np
```

```python
# read in the data
homeownership = pd.read_csv("Homeownership.csv")
neighborhood = pd.read_csv("Neighborhoods.csv")
# check types
print(homeownership.dtypes)
# fix types to be numeric
homeownership['TotalPopulation'] = homeownership['TotalPopulation'].str.replace(',', '').astype(float)
homeownership['OwnedMortgage'] = homeownership['OwnedMortgage'].str.replace(',', '').astype(float)
homeownership['OwnedFree'] = homeownership['OwnedFree'].str.replace(',', '').astype(float)
homeownership['RenterOccupied'] = homeownership['RenterOccupied'].str.replace(',', '').astype(float)
print(homeownership.dtypes)
homeownership.head(10)
```

I am going to add OwnedMortgage and OwnedFree columns. Then, I am going to find the census tract(s) with the largest ratio of Owner/Renting. After finding that, I will use the neighborhoods dataset to map the census tract to the neighborhoods.

```python
neighborhood.sample(1)
neighborhood['tractce10'] = neighborhood['tractce10'].str.replace('','').astype(float, errors = 'ignore')

#
#census_tract = neighborhood['tracece10']; # convert to list
census_tract = pd.Series(neighborhood.iloc[:,4])
for i in range(len(census_tract)):
    string = census_tract[i]
    if (string != ''):
        firstChar = string[0]
        if (firstChar == '0'):
            # if the first number (index 0) of the tract is 0, remove that character
            string = string[1:]
        # remove extra zeros
        census_tract[i] = string[0:len(string)-2]
# create data frame and cast strings to float
censusTracts = pd.DataFrame(census_tract).astype(float, errors = 'ignore')
print(censusTracts.dtypes)
censusTracts.sample(10)
```

```python
# add number of houses with mortgage and houses with mortgage payed off
mortgageHolders = homeownership['OwnedMortgage']
houseOwners = homeownership['OwnedFree']
homeOwners = mortgageHolders + houseOwners
# find ratio of renters because renting is associated with short term/transitional housing while owning a house/paying mortgage is
#^ characteristic of people (most likely families) settling down
renters = homeownership['RenterOccupied']
ratioSettled2Renting = homeOwners/renters
ratioSettled2Renting.head(4)
```

```python
# relate back to the census track (create/merge new dataframes)
ratio = pd.DataFrame(ratioSettled2Renting, columns=["Settled/renting"])
census_tracts = pd.DataFrame(homeownership['Census Tract'])
population = pd.DataFrame(homeownership['TotalPopulation'])

df3 = census_tracts.join(ratio).join(population)
df3.head(10)
```

```python
# sort (we want the census tracts with the highest amount of house owners to renters)
dfsorted = df3.sort_values(by=['Settled/renting'], ascending=False)
# get rid of any tract with less than 1000 people
dfFiltered = dfsorted[dfsorted['TotalPopulation'] >= 500]

dfFiltered.head(10)
```
