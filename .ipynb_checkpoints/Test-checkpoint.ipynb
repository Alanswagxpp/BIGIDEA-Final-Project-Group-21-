```python
import pandas as pd
import numpy as np
import math
%matplotlib inline
import matplotlib.pyplot as plt
import geopandas
```

```python
sideWalk = pd.read_csv("Sidewalk Ratio.csv")
neighborhood = pd.read_csv("neighborhood.csv")

sideWalk = sideWalk.loc[:, ['GEOID', 'Ratio']]
neighborhood = neighborhood.loc[:, ['geoid10', 'hood']]

# make a mask to delete NaN Ratio neighborhood
sideWalk_mask = sideWalk['Ratio'] >= 0
sideWalk = sideWalk[sideWalk_mask]

sideWalk.head(5)
```

```python
neighborhood.rename(columns = {'geoid10':'GEOID'}, inplace = True)
neighborhood.head(5)
```

```python
# utilize Pandas Series to convert all elements into one datatype
geoid = pd.Series(neighborhood.iloc[:,0])

for i in range(len(geoid)):
    string =geoid[i]
    # if not empty string
    if (string != ''):
        firstChar = string[0]
        try:
            float(firstChar)
            # if the firstChar is a number that starts with 0 as a placeholder (e.g., 024500)
            if (firstChar == '0' and len(string) > 1):
                # remove the first character
                string = string[1:]
            # remove extra zeros at the end
            geoid[i] = string[0:len(string)-2]
            
        # first char cannot be converted to a float
        except ValueError:
            geoid[i] = 0
    
    # if empty string
    else:
        geoid[i] = 0
# end of for loop

geoid = pd.DataFrame(geoid).astype(float, errors = 'raise')
geoid = geoid/100000
nbhd = geoid.join(neighborhood['hood'])
nbhd
```

```python
#sw = #truncate
sw = sideWalk['GEOID']/10000000
sw = pd.DataFrame(sw)
sw = sw.join(sideWalk['Ratio'])
sw
```

```python
sw_ratio = [0]*len(nbhd)
n = 0
for i in nbhd['GEOID']:
    # find i in Sidewalk Ratio dataset
    sw_ratio[n] = 0
    cnt = 0
    for j in sw['GEOID']:
        #j = j-j*1000
        #print(j)
        diff = abs(i-j)
        if diff <= 0.00001:
            sw_ratio[n] = sw['Ratio'][cnt]
            break
        cnt = cnt + 1
    n = n + 1
```

```python
# create dataframe and cast elements to float
sw_ratio_df = pd.DataFrame(sw_ratio)
nbhd_sideWalkRatio = pd.DataFrame(nbhd['hood']).join(sw_ratio_df)
nbhd_sideWalkRatio.columns =['Neighborhood', 'Sidewalk/Street']
nbhd_sideWalkRatio = nbhd_sideWalkRatio.sort_values(by=['Sidewalk/Street'], ascending=False)
nbhd_sideWalkRatio.head(20)
```
